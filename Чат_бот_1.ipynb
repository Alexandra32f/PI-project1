```python
from transformers import pipeline

# загрузка модели для машинного обучения
qa_pipeline = pipeline('question-answering', model='distilbert-base-uncased-distilled-squad', tokenizer='distilbert-base-uncased')

# функция для обработки вопросов и получения ответов от модели
def chat_bot(question):
    answer = qa_pipeline({
        'question': question,
        'context': context
    })
    return answer['answer']

# пример использования
context = input("Введи context: ")
question = input("Введи вопрос: ")
answer = chat_bot(question)
print(answer)
```
